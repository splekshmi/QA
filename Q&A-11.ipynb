### Reading the dataset to python environment

```python
import pandas as pd
Student = pd.read_csv('student-data.csv')
```

Dataset consist of  395 rows and 31 columns

Exploring the dataset to determine how many students we have information on, and learn about the graduation rate among these students.

```python
students = len(Student.index)
passed = len(Student[(Student.passed == 'yes')])
failed = len(Student[(Student.passed == 'no')])
grad_rate = float(passed)/float(students)*100.0

# Print the results
print ("Total number of students: {}".format(students))
print ("Number of students who passed: {}".format(passed))
print ("Number of students who failed: {}".format(failed))
print ("Graduation rate of the class: {:.2f}%".format(grad_rate))
```

Since the current graduation rate is 67.09% and the intended rate after ten years is 95%,timely intervention is necessary.

#### Preprocessing the dataset

```python
#before preprocessing identifying columns that require encoding 
for i in Student.columns:
    print(i,Student[i].unique(),Student[i].dtype)
```

```python
# Converting all categorical columns to numerical columns 
```

```python
Student.replace(['yes', 'no'], [1, 0],inplace =True) 
```

```python
for x in Student.columns:
        if Student[x].dtype==object:
            dummies = pd.get_dummies(Student[x], prefix=x, dummy_na=False)
            Student = Student.drop(x,axis=1)
            Student = pd.concat([Student, dummies], axis=1)
```

```python
Student.info()
```

#### Feature Engineering

```python
# Adding up mothers education and fathers education to Parents education,denoted by Pedu
Student['Pedu'] = Student['Medu'] + Student['Fedu']
Student.drop(['Medu','Fedu'],axis=1)
```

#### Building the model and evaluating the performance

```python
# spliiting the dataset into features and target
#taking 'passed' as target(y) and all other as X
X=Student.drop(['passed'],axis=1)
y=Student['passed']
```

```python
# splitting into test and training set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.20)
```

```python
# checking how many data points are there in the training set
X_train.shape
```

316 datapoints are given for training and 49 data points for testing

```python
print('Grad rate of train set:{:.2f}%'.format(100 * (y_train == 1).mean()))
print('Grad rate of test set: {:.2f}%'.format(100 * (y_test == 1).mean()))
```

```python
# importing the required libraries
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
import time
from sklearn.metrics import f1_score
```

```python
print ("Training using LogisticRegression with a training set size of ", len(X_train))
logit_model=LogisticRegression(random_state=42,max_iter=10000)
start =time.time()
logit_model.fit(X_train,y_train)
end = time.time()
print ("Trained model in {:.4f} seconds".format(end - start))
start=time.time()
y_pred=logit_model.predict(X_test)
end=time.time()
print ("Made predictions in {:.4f} seconds".format(end - start))
print("f1 score is:",round(f1_score(y_test,y_pred),3))
```

Other than computation timing ,F1 score is used to measure the performance of the model

F1 score is the harmonic mean of Precision and Recall and gives a better measure than the Accuracy Metric.

```python
print ("Training using SVM model with a training set size of ", len(X_train))
svm_model=SVC(random_state=42)
start =time.time()
svm_model.fit(X_train,y_train)
end = time.time()
print ("Trained model in {:.4f} seconds".format(end - start))
start =time.time()
y_pred=svm_model.predict(X_test)
end=time.time()
print ("Made predictions in {:.4f} seconds".format(end - start))
print("f1 score is:",round(f1_score(y_test,y_pred),3))
```

```python
print ("Training using KNN with a training set size of ", len(X_train))
knn_model=KNeighborsClassifier()
start =time.time()
knn_model.fit(X_train,y_train)
end = time.time()
print ("Trained model in {:.4f} seconds".format(end - start))
start =time.time()
y_pred=knn_model.predict(X_test)
end=time.time()
print ("Made predictions in {:.4f} seconds".format(end - start))
print("f1 score is:",round(f1_score(y_test,y_pred),3))
```

```python
print ("Training using Random Forest with a training set size of ", len(X_train))
random_forest=RandomForestClassifier(random_state=42)
start =time.time()
random_forest.fit(X_train,y_train)
end = time.time()
print ("Trained model in {:.4f} seconds".format(end - start))
# predicing the values using X_test by the model created
start =time.time()
y_pred=random_forest.predict(X_test)
end=time.time()
print ("Made predictions in {:.4f} seconds".format(end - start))
print("f1 score is:",round(f1_score(y_test,y_pred),3))
```

```python
print ("Training using Decision tree with a training set size of ", len(X_train))
decision_tree=DecisionTreeClassifier(random_state=42)
start =time.time()
decision_tree.fit(X_train,y_train)
end = time.time()
print ("Trained model in {:.4f} seconds".format(end - start))
start =time.time()
y_pred=decision_tree.predict(X_test)
end=time.time()
print ("Made predictions in {:.4f} seconds".format(end - start))
print("f1 score is:",round(f1_score(y_test,y_pred),3))
```

Based on the computation cost(including training time and prediction time) and F1_score, SVM performs better than other models.Since the data set is comparatively small ,SVM is the better option.
