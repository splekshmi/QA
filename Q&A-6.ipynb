### 2) Examining the data using visualisatons

##### Analysing polpulation across regions

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches       # for customizing legends
plt.figure(figsize=(12,12))
Ecno_data.groupby(['Region']).sum().plot(kind='bar', y='Population in thousands (2017)',legend=None,width=0.9,fontsize=10)
plt.title('Population across regions(Year:2017) ',fontsize=16)
plt.ylabel('Population in Millions',fontsize=14)
plt.xlabel('Region',fontsize=14)
leg_1 = mpatches.Patch(label='Population in Millions')  #using bbox_to_anchor for displaying the legends outside the graph
plt.legend(handles=[leg_1],bbox_to_anchor=(1,1))    
plt.show()
```

##### Analysing GDP per capita across regions (by average)

```python
plt.figure(figsize=(12,12))
round(Ecno_data.groupby(['Region']).mean(),2).plot(kind='bar', y='GDP per capita (current US$)',color='Red',legend=None,width=0.9,fontsize=10)
plt.title('GDP per capita across regions(by average)',fontsize=16)
plt.ylabel('GDP percapita(by average)',fontsize=14)
plt.xlabel('Region',fontsize=14)
leg_1 = mpatches.Patch(color='Red',label='GDP per capita(by average)')  #using bbox_to_anchor for displaying the legends outside the graph
plt.legend(handles=[leg_1],bbox_to_anchor=(1,1))    
plt.show()
```

##### Checking whether a higher agricultural production index contributes to higher GDP per capita

```python
plt.figure(figsize=(6,6)) 
plt.scatter(Ecno_data['Agricultural production index '],Ecno_data['GDP per capita (current US$)'],c='Red',s=12,marker='^')
plt.title('Agricultural production index Vs GDP per capita',fontsize=16)
plt.xlabel('Agricultural production index')  
plt.ylabel('GDP per capita(current US$)')
plt.show()
```

```python
##### Checking whether a higher population contributes to higher imports
```

```python
plt.figure(figsize=(8,8)) 
plt.scatter(Ecno_data['Population in thousands (2017)'],Ecno_data['International trade: Imports (million US$)'],s=12,marker='^')
plt.title('Populations Vs Imports',fontsize=16)
plt.xlabel('Population in Millions (Year:2017)')  
plt.ylabel('Imports (million US$)')
plt.show()
```

###  4(a) On average which region has the highest GDP per capita (current US$) 


```python
 # finding the average GDP percapita for every region
Ecno_data2=round(Ecno_data.groupby(['Region'])[['GDP per capita (current US$)']].mean(),2)  
Ecno_data2.set_axis(['GDP per capita(by average)'],axis=1)  # to give appropriate column name
```

```python
# Displaying the region with highest GDP percapita 
print("Region with highest average",Ecno_data2.idxmax())
```

###  4(b) On average which region has the highest International trade: Imports (million US$)

```python
Ecno_data3=round(Ecno_data.groupby(['Region'])[['International trade: Imports (million US$)']].mean(),2)  
Ecno_data3.set_axis(['International trade: Imports (million US$)(by average)'],axis=1)  # to give appropriate column name
```

```python
# Displaying the region with highest International trade: Imports (million US$) 
print("Region with highest average",Ecno_data3.idxmax())
```

#### Find out which region is more consistent with respect to GDP per capita (current US$)

```python
Ecno_data4=round(Ecno_data.groupby('Region')[['GDP per capita (current US$)']].std(),2)
```

```python
Ecno_data4.set_axis(['Standard deviation with respect to GDP per capita'],axis=1)
```

```python
#A smaller standard deviation means greater consistency
# Displaying the region which is more consistent with respect to GDP per capita
print("Region which is more consistent with ",Ecno_data4.idxmin())
```
