#### 1) Compare churn count with respect to gender

#### importing data set to python environment

```python
import pandas as pd
dc = pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
dc.head(5) # to get an overall idea about data set
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.countplot(x='gender',hue='Churn',data=dc)
plt.show()
```

#### 2) Find out how many female senior citizens there in the dataset

```python
import matplotlib.patches as mpatches
sc=dc['SeniorCitizen']==1.0
gc= dc['gender']=='Female'
dt1=dc.loc[sc & gc,:]
ax=sns.countplot(x='gender',hue='SeniorCitizen',data=dt1)
ax.bar_label(ax.containers[0])
plt.show()
```

#### 3) Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize=(6,6))
plt.scatter(dc['tenure'],dc['TotalCharges'],color='Red',s=5) # 's' denotes the size of marker (dot)
plt.title('Tenure VS Total Charges',fontsize=16)
plt.xlabel('Tenure')  
plt.ylabel('Total Charges')
plt.show()
```

#### 4) Find out which contract is preferred by the senior citizen

```python
plt.figure(figsize=(6,6))
dt2=dc[dc["SeniorCitizen"]==1.0]
plt.title(" Preference of Contract by Senior Citizens ", fontsize=16)
dt2.Contract.value_counts(sort=True).plot.pie(fontsize=12,startangle=90,autopct = '%1.1f')
plt.show()
```

#### 5) Comment on your finds on Payment Method

```python
plt.figure(figsize=(8,8))
sns.countplot(data =dc, x = 'PaymentMethod', hue=dc['gender'])
plt.title( "Preference of payment method:Female vs Male",fontsize=16) 
plt.xlabel('Payment Method', fontsize=14)
plt.ylabel('Count',fontsize=14) 
plt.tight_layout()
plt.show()
```

```python
# Both genders prefer Electronic check over other payment methods.
```

```python
plt.figure(figsize=(5,5))
dt3=dc[dc["SeniorCitizen"]==1.0]
plt.title(" Preference of payment method by Senior Citizens ", fontsize=16)
dt3.PaymentMethod.value_counts(sort=True).plot.pie(fontsize=12,startangle=0,autopct = '%1.1f')
plt.show()
```

```python
# Senior citizens mostly prefers Electronic check
```
